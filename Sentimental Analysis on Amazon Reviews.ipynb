```python
eg  = df["reviews.text"][50]
print(eg)
```

```python
tokens = nltk.word_tokenize(eg)
tokens[:10]
```

```python
tagged = nltk.pos_tag(tokens)
print(tagged)
```

```python

entities = nltk.chunk.ne_chunk(tagged)


print(entities)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('I am so happy!!!')
```

```python
sia.polarity_scores(eg)
```

```python
#Run the polarity Scores in the entire database
result = {}
for i , row in tqdm(df.iterrows() , total = len(df)):
    text = row['reviews.text']
    myid = row['Id']
    result[myid] = sia.polarity_scores(text)
```

```python
result
```

```python
df.head()
```

```python
vaders = pd.DataFrame(result).T
vaders = vaders.reset_index().rename(columns={'index': "Id"})
vaders = vaders.merge(df, how = 'left')
```

```python
#Now we have sentiment scores and meatdata

vaders.head()
```

```python
##  Plot vader results
ax = sns.barplot(data = vaders, x = 'reviews.rating', y = 'compound')
ax.set_title('Compound Score by Amazon Star Review')
plt.show()
```

```python
fig, axs = plt.subplots(1,3, figsize =(15,5))

sns.barplot(data = vaders, x = "reviews.rating", y = "pos", ax=axs[0])
sns.barplot(data = vaders, x = "reviews.rating", y = "neu", ax = axs[1])
sns.barplot(data = vaders, x = "reviews.rating", y = "neg", ax = axs[2])
axs[0].set_title("Positive")
axs[1].set_title("Neutral")
axs[2].set_title("Negative")
plt.tight_layout()
plt.show()
```
